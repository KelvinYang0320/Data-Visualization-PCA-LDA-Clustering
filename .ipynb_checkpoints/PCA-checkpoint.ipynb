# Project 2 - Data Visulization: PCA, LDA, Clustering
This project is designed for you to futher understand PCA, LDA, and Data Clustering as introduced in class by real implementations on some data sets. 

* Compute the pooled $d\times d$ covariance matrix $C=\frac{1}{n}\sum^3_{k=1}\sum^{n_k}_{i=1}(x^{(k)}_i)(x^{(k)}_i)^t$, where $u=\sum^3_{k=1}\sum^{15}_{i=1}x^{(k)}_i$ is the mean vector.

```python
import numpy as np
f = open("8OX/data8OX.txt", 'r')
print(f.read())
n = 3
data = np.loadtxt("8OX/data8OX.txt",skiprows=n)

data = data[:,:-1]
```

```python
def PCA(X, K, showDetailedInfo=False):
    assert str(type(X)) == "<class 'numpy.ndarray'>", 'X must be numpy.ndarray'
    [n, d] = X.shape
    # Compute the mean vector
    u = np.sum(X, axis=0)/n
    C = np.cov(X, rowvar=False)
    
    [L, U] = np.linalg.eig(C)
    sorted = -np.sort(-L)
    if (showDetailedInfo):
        print("Covariance Matrix:\n", C)
        print("Sorted Eigenvalues:\n", sorted)
        sumSorted = np.sum(sorted)
        cnt = 0
        percentageList = []
        for i in sorted:
            cnt = cnt + i
            percentageList.append(cnt/sumSorted)
        print("percentage:\n", percentageList)
    index = np.argsort(-L)
    Xproj = np.zeros((d, K))
    for i in range(K):
        Xproj[:, i] = U[:, index[i]]
    Y=np.matmul(X, Xproj)
    return Y
```

```python
import matplotlib.pyplot as plt
k = 2
Y = PCA(data, k, True)
X1 = -Y[0:15, 0]
X2 = -Y[16:30, 0]
X3 = -Y[30:45, 0]
Y1 = -Y[0:15, 1]
Y2 = -Y[16:30, 1]
Y3 = -Y[30:45, 1]

plt.figure()
plt.title("Firse Two Principal Component Projection for 8OX Data")
plt.scatter(X1,Y1, marker="d", c="w", edgecolors='c', s=100, label="8") #x o
plt.scatter(X2,Y2, marker="o", c="w", edgecolors='r', s=100, label="O") #x o
plt.scatter(X3,Y3, marker="x", c="y", s=100, label="X") #x o
plt.legend()
```
